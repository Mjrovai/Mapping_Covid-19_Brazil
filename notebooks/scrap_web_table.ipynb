```python
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/States_of_Brazil'
html = requests.get(url).content
```

```python
df_list = pd.read_html(html)
```

```python
df = df_list[1]
df
```

```python
df.columns
```

```python
df.columns = ['state', 'UF', 'Capital', 'Area(km2)',
       'Population(2019)', 'Density (perkm2, 2019)',
       'GDP (R$millions, 2016)', 'HDI(2017)']
```

```python
df.head(2)
```

```python
df.to_csv('../data/brazil_states_info.csv')
```
